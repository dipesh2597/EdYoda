# Time Complexity:
Time taken by a program in execution.

We will take our own unit by name unit.

#### Program:
x = 1  ===> 1 unit of time

x = 1     |
          | ==> 1 unit of time
x = x+1   | 


for i in range(0,5):
    print(i)    ==> 1 unit of time
    
time complexity of loop in 5 unit.

for i in range(0,5):
    for j in range(0,5):
        print(j) ==> 1 unit of time
        
time complexity of loop in 5^2 unit.

for i in range(0,n):
    print(i)    ==> 1 unit of time
    
time complexity of loop in n unit. O(n) ~= 5n, 10n, 1million*n, 2n+3

for i in range(0,n):
    for j in range(0,n):
        print(j) ==> 1 unit of time
        
time complexity of loop in n^2 unit. O(n^2)

f(n) = 2n + n^2 + n^3 + 10 => O(n^3)

## Time Complexity Function:

O(1) < O(logn) < O(sqrt(n))  < O(n) < O(nlogn) < O(n^2) < O(n^3) < ...... < O(n^n) 

### Big O
    Upper bound of time complexity
    f(n) <= c*g(n)      ~= n <= n not 
    
    f(n) = 5n + 15 = O(n)
    g(n) = 20n
    
    n = 1, f(n) = 20, g(n) = 20
    n = 2, f(n) = 25, g(n) = 40
    
    
### Omega (music listening)
    Lower bound of time complexity
    f(n) >= c*g(n)
    
    
### Theta (degree celcius sign) 
    asymptomatic notation in middle of lower and upper
    c1*g(n) <= f(n) <= c2*g(n)
    5*sqrt(n) <= 5n + 15 <= 20n^2
