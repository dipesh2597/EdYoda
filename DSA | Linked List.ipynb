```python
# Element class
class Node:
    
    def __init__(self,data):
        self.data = data
        self.next = None
        
class LinkedList:
    
    def __init__(self):
        self.head = None
        self.tail = None
            
    def iterate_linked_list(self):
        current_node = self.head
        while current_node: # while current_node != None
            yield current_node
            current_node = current_node.next
            
    def add_node(self, node):
        if not self.head:
            print(f"Linked list is empty so adding {node.data} as first and last element")
            self.head = node
            self.tail = node
        else:
            print(f"Linked list is not empty so adding {node.data} after {self.tail.data}")
            self.tail.next = node
            self.tail = node
        
    def remove_node(self):
        if not self.head:
            print("Linked list is empty so deletion can't be performed")
        else:
            current_node = self.head
            prev_node = self.head
            
            while current_node:
                print("current node: ",current_node.data, "prev node: ", prev_node.data)
                if current_node.next:
                    prev_node = current_node
                print("AFTER UPDATING: ", "current node: ",current_node.data, "prev node: ", prev_node.data)
                
                current_node = current_node.next
            print("AFTER WHILE LOOP: ", "current node: ",current_node, "prev node: ", prev_node.data) 
            prev_node.next = None
            self.tail = prev_node
        

linked_list = LinkedList()

# linked list is empty so head would None
print(linked_list.head)
print("printing empty list")
for node in linked_list.iterate_linked_list():
    print(node.data)
node1 = Node(1)
linked_list.add_node(node1)

print("printing list after adding head")
for node in linked_list.iterate_linked_list():
    print(node.data)
    
node2 = Node(2)
node3 = Node(3)


# adding nodes in linked list one by one
#inserting
linked_list.add_node(node2)
print("tail after adding 2: ",linked_list.tail.data)   
linked_list.add_node(node3)


print("printing list after adding all three nodes")
for node in linked_list.iterate_linked_list():
    print(node.data)
    
#deletion 
linked_list.remove_node()
print("After removal tail is", linked_list.tail.data)
print("printing list after removing the last element")
for node in linked_list.iterate_linked_list():
    print(node.data)
```
