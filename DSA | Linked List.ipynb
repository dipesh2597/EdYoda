```python
# Element class
class Node:
    
    def __init__(self,data):
        self.data = data
        self.next = None
        
class LinkedList:
    
    def __init__(self):
        self.head = None
        self.tail = None
            
    def iterate_linked_list(self):
        current_node = self.head
        while current_node: # while current_node != None
            yield current_node
            current_node = current_node.next
            
    def add_node(self, node):
        if not self.head:
            print(f"Linked list is empty so adding {node.data} as first and last element")
            self.head = node
            self.tail = node
        else:
            print(f"Linked list is not empty so adding {node.data} after {self.tail.data}")
            self.tail.next = node
            self.tail = node
        
    def remove_node_from_end(self):
        if not self.head:
            print("Linked list is empty so deletion can't be performed")
        else:
            current_node = self.head
            prev_node = self.head
            
            while current_node:
                if current_node.next:
                    prev_node = current_node
                current_node = current_node.next
            prev_node.next = None
            self.tail = prev_node
            
    def remove_node_from_beggining(self):
        if not self.head:
            print("Linked list is empty so deletion can't be performed")
        else:
            current_head = self.head
            new_head_node = current_head.next
            current_head.next = None
            self.head = new_head_node
            
    def remove_node_from_middle(self, data):
        if not self.head:
            print("Linked list is empty so deletion can't be performed")
        else:
            current_node = self.head
            prev_node = self.head
            next_node = None
            
            while current_node:
                if current_node.data == data:
                    next_node = current_node.next
                    current_node.next = None
                    prev_node.next = next_node
                    break
                    
                if current_node.next:
                    prev_node = current_node
                    
                current_node = current_node.next
            
            
'''
node = |data|next|

node1 = |3|node2|  #connected to node2
node2 = |4|Null/None| # Not connected to any node

#disconnet node1 from node2
node1 = |3|None|

'''        

linked_list = LinkedList()

# linked list is empty so head would None
print(linked_list.head)
print("printing empty list")
for node in linked_list.iterate_linked_list():
    print(node.data)
node1 = Node(1)
linked_list.add_node(node1)

print("printing list after adding head")
for node in linked_list.iterate_linked_list():
    print(node.data)
    
node2 = Node(2)
node3 = Node(3)


# adding nodes in linked list one by one
#inserting
linked_list.add_node(node2)
print("tail after adding 2: ",linked_list.tail.data)   
linked_list.add_node(node3)


print("printing list after adding all three nodes")
for node in linked_list.iterate_linked_list():
    print(node.data)
    
#deletion from end 
linked_list.remove_node_from_end()
print("printing list after removing the last element")
for node in linked_list.iterate_linked_list():
    print(node.data)
    
#deletion from beginning
linked_list.remove_node_from_beggining()
print("printing list after removing the first element")
for node in linked_list.iterate_linked_list():
    print(node.data)
    
    
linked_list.add_node(node3)
node4 = Node(4)
linked_list.add_node(node4)
linked_list.add_node(node1)
print("printing new list")
for node in linked_list.iterate_linked_list():
    print(node.data)
    
#deletion from midlle
linked_list.remove_node_from_middle(4)
print("printing list after removing the 4 element")
for node in linked_list.iterate_linked_list():
    print(node.data)
```
