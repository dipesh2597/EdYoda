```python
from bs4 import BeautifulSoup
import requests
url = 'https://www.ayushsharma.in/technology'

# response = requests.request("get",url)
response = requests.get(url)

response_data=response.text
# print(response_data)
html_data = BeautifulSoup(response_data, 'html.parser')
print(type(html_data))

#select method to select a tag
# print(html_data.select('a'))

#select a tag by class name
print(type(html_data.select('a.post-card')))
a_post_card = html_data.select('a.post-card')

result = [] # [{'title':'vlaues','text':"value"}]
for pc in a_post_card:
    #select by class so we have to use . before classname
    data = {
        'title':pc.select('.card-title')[0].get_text(),
        'text':pc.select('.card-text')[0].get_text()
    }
    print(data)
    result.append(data)
    
print(result)
```
