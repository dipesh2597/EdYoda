```python
class Stack:
    
    def __init__(self):
        self.stack = []
        
    def push(self, element):
        self.stack.append(element)
        print(f"Element: {element} pushed in the stack")
    
    def is_empty(self):
#         return len(self.stack) == 0
        if len(self.stack) == 0:
            print("stack is empty")
        else:
            print("stack is not empty")
    
    def pop(self):
        if len(self.stack) == 0:
            print("POP can't be performed on empty stack")
        else:
            poped_element = self.stack.pop()
            print(f"Poped: {poped_element} from the stack")
        
            
    def size(self):
        print("The size of stack is: ",len(self.stack))
    
    def display(self):
        for i in self.stack:
            print(i)
            
    def peek(self):
        if len(self.stack) == 0:
            print("Peek can't be performed on empty stack")
        else:
            
            print(f"Top element is: {self.stack[-1]}")
        
            
my_stack = Stack()
my_stack.is_empty()
my_stack.pop()
my_stack.push(2)
my_stack.push(3)
my_stack.push(4)
my_stack.push(5)
my_stack.push(6)
my_stack.push(7)
my_stack.size()
my_stack.display()
my_stack.peek()
print("Pooping the top element")
my_stack.pop()

my_stack.size()
my_stack.display()

my_stack.peek()
```

```python
"""
WE have to implement a max_stack.
Requirements:
    1. We have to maintain 2 stack one is normal second is max_stack
    2. Max stack will be having only maximum value from the normal stack after each push and pop operations
    
max_stack.peek() should return current maximum value of normal stack
Normal stack.       Max Stack
|1| 
|5|                  |5|
|0|                  |1|
|1|

Poping

|5|                  |5|
|0|                  |1|
|1|


|0|                  |1|
|1|
"""

class Stack:
    
    def __init__(self):
        self.stack = []
        self.max_stack = []
        
    def push(self, element):
        self.stack.append(element)
        if len(self.max_stack) == 0:
            self.max_stack.append(element)
        elif self.max_stack[-1] <= element:
            self.max_stack.append(element)
        
        print(f"Element: {element} pushed in the stack")
    
    def is_empty(self):
#         return len(self.stack) == 0
        if len(self.stack) == 0:
            print("stack is empty")
        else:
            print("stack is not empty")
    
    def pop(self):
        if len(self.stack) == 0:
            print("POP can't be performed on empty stack")
        else:
            poped_element = self.stack.pop()
            if self.max_stack[-1] == poped_element:
                self.max_stack.pop()
            print(f"Poped: {poped_element} from the stack")
        
            
    def size(self):
        print("The size of stack is: ",len(self.stack))
    
    def display_normal_stack(self):
        print("Normal Stack")
        for i in self.stack[::-1]:
            print(f"|{i}|")
            
    def display_max_stack(self):
        print("Max Stack")
        for i in self.max_stack[::-1]:
            print(f"|{i}|")
            
    def peek(self):
        if len(self.stack) == 0:
            print("Peek can't be performed on empty stack")
        else:
            
            print(f"Top element is: {self.stack[-1]}")
        
    def get_max(self):
        print("Maximum is: ",self.max_stack[-1])
        

my_stack = Stack()
my_stack.push(1)
my_stack.display_normal_stack()
my_stack.display_max_stack()
my_stack.push(0)
my_stack.display_normal_stack()
my_stack.display_max_stack()
my_stack.push(5)
my_stack.display_normal_stack()
my_stack.display_max_stack()
my_stack.get_max()
my_stack.push(1)
my_stack.display_normal_stack()
my_stack.display_max_stack()
my_stack.get_max()
my_stack.pop()
my_stack.display_normal_stack()
my_stack.display_max_stack()
my_stack.get_max()
my_stack.pop()
my_stack.display_normal_stack()
my_stack.display_max_stack()
my_stack.get_max()
```
