```python
class Movie():
    
    def __init__(self,rows,columns):
        self.rows = rows
        self.columns = columns
        self.booking = {}
#         self.ticket_booked = 0
        
        '''
            1 2 3 4 5
        1.  S S S S S
        2.  S S S S S
        '''
    def is_booked(self,row,col):
        #logic
        seat_id = str(row)+str(col)
        if seat_id in self.booking:
            return True
        else:
            return False
        
    def show_seats(self):
        print("Here is seat allocation available for you: ")
        for row in range(0,self.rows+1):
            for column in range(0,self.columns+1):
                if row == 0:
                    if column == 0:
                        print(" ", end=" ")
                    else:
                        print(column, end=" ")
                else:
                    if column == 0:
                        print(row, end=" ")
                    else:
                        # if check for booked row and columns print("B") else print("S")
                        if self.is_booked(row,column):
                            print("B", end=" ")
                        else:
                            print("S", end=" ")
            print()
        
    def book_ticket(self):
        print("Ticket Counter")
        booking_row = int(input("Enter number of row you want to book: "))
        booking_column = int(input("Enter number of column you want to book: "))
        seat_id = str(booking_row)+str(booking_column)
        #here is your price: Pricing rules
        no_of_seats = self.rows*self.columns
        if no_of_seats <= 60:
            price = 10
        else:
            row_factor = self.rows//2
            if booking_row <= row_factor:
                price = 10
            else:
                price = 8
        print("Here is the price of your ticket: ",price)
        choice = int(input("Do you want to book?\n1. Yes\n2. No\nEnter your choice: "))
        if choice == 1:
            print("We're booking your ticket:")
            name = input("Please enter your name: ")
            gender = input("Please enter your gender: ")
            age = input("Please enter your age: ")
            mob = input("Please enter your mobile number: ")
            details = {'name':name,'gender':gender,'age':age,'mob':mob,'price':price}
            self.booking[seat_id]=details
#             self.ticket_booked += 1
            print("Ticket Booked Successfully!!!")
        elif choice == 2:
            print("Cancelled Booking")
            
    def show_details(self,row,col):
        seat_id = str(row)+str(col)
        if seat_id in self.booking:
            detail = self.booking[seat_id]
            print("Here is the deatils: ")
            print("Name: ",detail['name'])
            print("Age: ",detail['age'])
            print("Gender: ",detail['gender'])
            print("Mobile number: ",detail['mob'])
            print("Ticket book for : $",detail['price'])
        else:
            print(f"Seat for row: {row} and column: {col} is not booked yet")
            
    def statics(self):
        booked_ticket = len(self.booking)
        print("Number of ticket purchased: ",booked_ticket)
        total_ticket = self.rows*self.columns
        percentage_of_ticket_booked = round((booked_ticket/total_ticket)*100,2)
        print("Percentage of ticket booked: ",percentage_of_ticket_booked)
        current_income = 0
        for seat_id in self.booking:
            details = m1.booking[seat_id]
            price = details['price']
            current_income += price
        print("Current income: ",current_income)
        if total_ticket <= 60:
            total_income = total_ticket*10
        else:
            row_factor = self.rows//2
            front_rows_tickets = row_factor * self.columns
            back_rows_tickets = (self.rows-row_factor)*self.columns
            total_income = front_rows_tickets*10 + back_rows_tickets*8
        print("Total income: ",total_income)
```
